```python
import pandas as pd
### Loc_Pt
## create minType table
# Load your data
df = pd.read_csv('/Users/nguyenngoc/Documents/CSE385 /Final Project/2NF/Loc_pt/siteLocPt.csv')

# Step 1: Normalize by extracting unique rows based on the 'Type' column
mint = df[['Ftr_Type', 'Ftr_Group']].drop_duplicates().reset_index(drop=True)

# Step 2: Create a 'Type_ID' column with the syntax 'T_0' followed by counting number
mint['Type_ID'] = 'T_' + (mint.index + 1).astype(str).str.zfill(2)

# Step 3: Reorder columns to place 'Type_ID' as the first column
mint = mint[['Type_ID', 'Ftr_Type', 'Ftr_Group']]

# Save the result to a CSV file
mint.to_csv('minTypeLoc.csv', index=False)
```

```python
## Creating the relations table 

# Load your data 
df1 = pd.read_csv('/Users/nguyenngoc/Documents/CSE385 /Final Project/1NF/1NF_Loc_Pt.csv')
# Specify the columns to include in the relation table
columns_to_include = ['Site_ID *', 'Ftr_ID *', 'Commodity', 'Ftr_Type']  

# Step 1: Create a new DataFrame with the specified columns
relation_Loc_Pt = df1[columns_to_include].drop_duplicates().reset_index(drop=True)

# Step 2: create a new column 'LOC_0' with the specified format
relation_Loc_Pt['LOCPT_ID'] = (
    'LT_0' + (relation_Loc_Pt.index + 1).astype(str)
)

# Step 3: Move this column to the first position
columns = ['LOCPT_ID'] + [col for col in relation_Loc_Pt.columns if col != 'LOCPT_ID']
relation_Loc_Pt = relation_Loc_Pt[columns]

# Step 3: Save or display the relation table
relation_Loc_Pt.to_csv('relation_Loc_Pt.csv', index=False)
print(relation_Loc_Pt)
```

```python
## Loc_Poly
# Creating the relations table 
df2 = pd.read_csv('/Users/nguyenngoc/Documents/CSE385 /Final Project/USGS_Cobalt_US_CSV/Loc_Poly.csv')
# Specify the columns to include in the relation table
columns_to_include = ['Site_ID', 'Ftr_ID']  

# Step 1: Create a new DataFrame with the specified columns
relation_Loc_Poly = df2[columns_to_include].drop_duplicates().reset_index(drop=True)

# Step 2: create a new column 'SiteFtCom_ID' with the specified format
relation_Loc_Poly['SiteFt_ID'] = (
    'SF_0' + (relation_Loc_Poly.index + 1).astype(str)
)

# Step 3: Move this column to the first position
columns = ['SiteFt_ID'] + [col for col in relation_Loc_Poly.columns if col != 'SiteFt_ID']
relation_Loc_Poly = relation_Loc_Poly[columns]

# Step 4: Save or display the relation table
relation_Loc_Poly.to_csv('relation_Loc_Poly.csv', index=False)
```
