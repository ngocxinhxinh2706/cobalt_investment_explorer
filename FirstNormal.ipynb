```python
import pandas as pd

# Load Loc_Pt file into DataFrame
df = pd.read_csv('/Users/nguyenngoc/Documents/CSE385 /Final Project/USGS_Cobalt_US_CSV/Loc_Pt.csv')
# Split and explode the multivalued column
df = df.assign(Commodity=df['Commodity'].str.split(';')).explode('Commodity')

# Save the normalized DataFrame back to a CSV file
df.to_csv('1NF_Loc_Pt.csv', index=False)
```
